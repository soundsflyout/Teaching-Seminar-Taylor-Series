```python
import numpy as np 
import math 
import matplotlib.pyplot as plt 
import ipywidgets as widgets

def cos_approximation_plot(degree, xlimit = 6, precision = 10):
    if precision <= 0 or xlimit <= 0:
        return print("Error: values must be positive")

    taylor_approximate_cos = lambda degree, x: sum([((-1)**(k//2))*x**k/math.factorial(k) for k in range(0,degree+1,2)])

    x_vals = [x/precision for x in range(-xlimit*precision,xlimit*precision)]

    plt.plot(x_vals, [taylor_approximate_cos(degree,x) for x in x_vals], 'r')
    plt.plot(x_vals, [math.cos(x) for x in x_vals], 'b')
    plt.ylim(-2,2)
    plt.show()

def exp_approximation_plot(degree, xlimit = 6, precision = 7):
    if precision <= 0 or xlimit <= 0:
        return print("Error: values must be positive")

    taylor_approximate_exp = lambda degree, x: sum([x**k/math.factorial(k) for k in range(0, degree+1)])  

    x_vals = [x/precision for x in range(-2*precision,xlimit*precision)]
    
    plt.plot(x_vals, [taylor_approximate_exp(degree,x) for x in x_vals], 'r')
    plt.plot(x_vals, [math.exp(x) for x in x_vals], 'b')
    plt.ylim(-100,math.exp(xlimit))
    plt.show()

def log_approximation_plot(degree, xlimit = 3, precision = 20):
    if precision <= 0 or xlimit <= 0:
        return print("Error: values must be positive")

    taylor_approximate_log = lambda degree, x: sum([((-1)**(k+1))*(x**k)/k for k in range(1,degree+1)])

    x_vals = [x/precision for x in range(-precision+1,xlimit*precision)]
    plt.plot([1+x for x in x_vals], [taylor_approximate_log(degree,x) for x in x_vals], 'r')
    plt.plot([1+x for x in x_vals], [math.log(1+x) for x in x_vals], 'b')
    plt.ylim(-3,math.log(xlimit)+2)
    plt.show()
```

```python
widgets.interact(cos_approximation_plot, degree=widgets.IntSlider(min=0, max=40, step=2, value=0))
```
